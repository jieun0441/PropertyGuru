```python
# if there's Condominium, Condominium column =1 and otherwise 0
def new_col_creating(df_name):
  result = df_name['Type'].apply(lambda x : 'Condominium' in str(x))
  result = result.astype('int')
  return result

df['Condo'] = new_col_creating(df)
df[['Type','Condo']]
```

```python
# if there's HDB, HDB column =1 and otherwise 0
def new_col_creating(df_name):
  result = df_name['Type'].apply(lambda x : 'HDB' in str(x))
  result = result.astype('int')
  return result

df['HDB'] = new_col_creating(df)
df[['Type','HDB', 'Condo']].tail(15)
```

```python
temp = list(range(df.shape[0]))
def new_col_creating(df_name):
	for i in range(df_name.shape[0]): 
		if df_name.iloc[i]['Condo']==1:
			temp[i] = 'Condo'
		elif df_name.iloc[i]['HDB']==1:
			temp[i] ='HDB'
		else:
			temp[i] ='other'
	return temp

df['House_type'] = new_col_creating(df)
```

### 팬데믹Phase 시간대 구분 변수 생성


```python
# 앞(Month파트)에서 시간변수로 데이터타입 변경 코드 돌린 다음에 진행하여야 함
# 해당 코드: df.listed_date = pd.to_datetime(df.listed_date)
# 일단 phase라는 새로운 변수를 생성
df['phase'] = 'None'
df[['listed_date', 'phase']]
```

```python
# phase cuts
import datetime
df.loc[(df['listed_date'] <=datetime.datetime(2020,1,23)), 'phase'] = '1) before covid'
df.loc[(df['listed_date'] >=datetime.datetime(2020,1,23)) & (df['listed_date'] <=datetime.datetime(2020,4,6) ), 'phase'] = '1.5) covid_early_stage'
df.loc[(df['listed_date'] >=datetime.datetime(2020,4,7)) & (df['listed_date'] <=datetime.datetime(2020,6,1) ), 'phase'] = '2) Circuit breaker lockdown'
df.loc[(df['listed_date'] >=datetime.datetime(2020,6,2)) & (df['listed_date'] <=datetime.datetime(2020,6,18) ), 'phase'] = '3) phase1'
df.loc[(df['listed_date'] >=datetime.datetime(2020,6,19)) & (df['listed_date'] <=datetime.datetime(2020,12,27) ), 'phase'] = '4) phase2'
df.loc[(df['listed_date'] >=datetime.datetime(2020,12,28)) & (df['listed_date'] <=datetime.datetime(2021,5,7) ), 'phase'] = '5) phase3'
df.loc[(df['listed_date'] >=datetime.datetime(2021,5,8)) & (df['listed_date'] <=datetime.datetime(2021,5,15) ), 'phase'] = '6) phase3(Heightened Alert)'
df.loc[(df['listed_date'] >=datetime.datetime(2021,5,16)) & (df['listed_date'] <=datetime.datetime(2021,6,13) ), 'phase'] = '7) phase2(Heightened Alert)'
df.loc[(df['listed_date'] >=datetime.datetime(2021,6,14)) & (df['listed_date'] <=datetime.datetime(2021,6,20) ), 'phase'] = '8) phase3(Heightened Alert) with Stage1'
df.loc[(df['listed_date'] >=datetime.datetime(2021,6,21)) & (df['listed_date'] <=datetime.datetime(2021,7,11) ), 'phase'] = '9) phase3(Heightened Alert) with Stage2'
df.loc[(df['listed_date'] >=datetime.datetime(2021,7,12)) & (df['listed_date'] <=datetime.datetime(2021,7,21) ), 'phase'] = '10) phase3(Heightened Alert)'
df.loc[(df['listed_date'] >=datetime.datetime(2021,7,22)) & (df['listed_date'] <=datetime.datetime(2021,8,9) ), 'phase'] = '11) phase2(Heightened Alert)'
```

```python
df[['listed_date','phase']].sample(20)
```

### question

```python
# 시간대 구분함수 만들기
def cut(x):
    if x <= '2020-01-23':
        return  'before covid'
    if '2020-04-07' <= x <= '2020-06-01':
        return 'circuit breaker'
    if '2020-06-02' <= x <= '2020-06-18':
        return 'phase1'
    if '2020-06-19' <= x <= '2020-12-27':
        return 'phase2'
    if '2020-12-28' <= x <= '2021-05-07':
        return 'phase3'
    if '2021-05-08' <= x <= '2021-05-15':
        return 'phase3(Heightened Alert)'
    if '2021-05-16' <= x <= '2021-06-13':
        return 'phase2(Heightened Alert)'
    if '2021-06-14' <= x <= '2021-06-20':
        return 'phase3(Heightened Alert) with Stage 1'
    if '2021-06-21' <= x <= '2021-07-11':
        return 'phase3(Heightened Alert) with Stage 2'
    if '2021-07-12' <= x <= '2021-07-21':
        return 'phase3(Heightened Alert)'    
    if '2021-07-22' <= x <= '2021-08-09':
        return 'phase2(Heightened Alert)'  
```

```python
df['phase'] = df['listed_date'].apply(lambda x : cut(x))
df[['phase', 'listed_date']].sample(20)
```
