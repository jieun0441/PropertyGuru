```python
# Stopwords to remove from text will have little effect on the context of the text
# Stopwords in list already in lower case
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
# Custom_stopwords_for_WordCloud = ['enquire', 'developer sale', 'call enquire', 'call now', 'to enquire', 'for sale', 'new listing', 'unit sale', 'serious seller', 'exclusive listing', 'floor unit', 'new list', 'new exclusive', 'direct developer','don’t miss','call','now','to','for','new','listing','unit','serious','seller','exclusive','floor','list','direct','developer','sale','call', 'now', 'enquire', 'unit', 'floor', 'serious', 'seller', 'listing', '3','2','1', '4', '5', 'blk', 'facing','ang','mo','kio']
# Custom_stopwords_for_one_word =['enquire', 'now', 'call', 'unit', 'floor', 'serious', 'seller', 'listing', '3', '2', '1', '4', '5', 'blk', 'facing', 'ang','mo','kio']
# Custom_stopwords_for_bigrams =['call enquire', 'call now','to enquire', 'for sale', 'new listing', 'unit sale', 'serious seller', 'exclusive listing', 'floor unit', 'new list', 'new exclusive', 'direct developer', 'developer sale']
# Custom_stopwords_for_trigrams = ['new exclusive listing', 'new listing sale', 'sale new listing', 'ang mo kio', 'direct developer sales','developer sales team', 'exclusive new listing', 'call enquire new', 'brand new listing']
stop_words = stop_words #+ Custom_stopwords
print(stop_words)
```

```python
# Use this to deal with apostophes and abbreviation

def remove_apostrophe_abbrev(text):
    return re.sub('[^\w\s]','', text)

# Function to remove stopwords
def remove_stop_words(text):
    clean_text = []
    for word in text.split():
        if word.strip().lower() not in stop_words:
            clean_text.append(word)
            
    return " ".join(clean_text)
```

```python
from nltk.stem import WordNetLemmatizer 
# Init the Wordnet Lemmatizer
def lemma(text):
    lemmatizer = WordNetLemmatizer()
    word_list = nltk.word_tokenize(text)
    # print(word_list)
    # print(type)


    # Lemmatize list of words and join
    lemmatized_output = ' '.join([lemmatizer.lemmatize(w,'v') for w in word_list])
    return lemmatized_output
```

```python
# Functions for cleaning text

# Remove html tags, using regex is bad idea
def remove_html_tags(text):
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text()

# Remove links in text, http and https
# s? in regex means or s (case sensitive)
def remove_links(text):
    return re.sub('https?:\/\/\S+', '', text)

# Remove 's (possessive pronouns) from text
# Two kinds of apostophes found
def remove_possessive_pronoun(text):
    return re.sub("’s|'s", '', text)

# Remove between brackets and their contents
def remove_between_brackets(text):
    return re.sub('\([^]]*\)', '', text)

# Remove square brackets and their contents
def remove_between_square_brackets(text):
    return re.sub('\[[^]]*\]', '', text)

# Remove curly brackets and their contents
# Useful to remove any JavaScript scripts, though removed
# through html as above
def remove_between_curly_brackets(text):
    return re.sub('\{[^]]*\}', '', text)

def remove_n_space(text):
    return re.sub('\n', '', text)
    
def abc(text):
  word_list = nltk.word_tokenize(text)
  # print(word_list)
  # print(type(word_list))
  return word_list
def text1(text):
  text = remove_html_tags(text)
  list1 = []
  text = abc(text)
  list1.append(text[:10]) # 자를단어 갯수 변경 여기서
  flat_list = [item for sublist in list1 for item in sublist]
  # print(flat_list)
  output = ' '.join(w for w in flat_list)
  return output


def text_cleaner(text):
    text = remove_html_tags(text)
    text = remove_links(text)
    text = remove_possessive_pronoun(text)
    text = remove_apostrophe_abbrev(text)
    text = remove_between_brackets(text)
    text = remove_between_square_brackets(text)
    text = remove_between_curly_brackets(text)
    text = remove_n_space(text)
    text = text.lower()
    text = remove_stop_words(text)
    text = lemma(text)
    text = text1(text)
    # print(text)
    
    return text
```
