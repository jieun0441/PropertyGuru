```python
# 결측치 비율이 높은 컬럼 리스트를 추출
# 컬럼 확인 가능한 데이터프레임 만들기. 
count_null = pd.DataFrame()

# 변수 이름이 들어간 컬럼 생성
count_null['variables'] = df.columns.values.tolist()

# 컬럼별로 결측값 갯수 세어서 NaN이라는 컬럼에 넣기
count_null['NaN'] = df.isnull().sum().tolist()

# 결측치 비율 계산하기 % calculation by columns
count_null['Nanperc'] = 100*(count_null['NaN']/len(df))
count_null

# 50% 이상이 결측치인 변수들의 리스트 구하기
too_many_NaN = count_null[count_null['Nanperc']>50]
will_erase = list(too_many_NaN['variables'])
```

##### 중요변수 결측치인 것 제거

```python
df4.info()
```

```python
# 분석에 필수적인 변수가 결측인 경우 삭제
important_vars = ['Listed_on', 'image_urls', 'price', 'latitude', 'longitude']
# drop the rows when important_vars column has missing
df4 = df4.dropna(subset = important_vars)
df4.info() # 39,601행 --> 39,517행
```

#### days_past

```python
df4['Listed_on']
```

```python
# 숫자 부분만 추출
df4['Listed_on_num'] = df4['Listed_on'].str.split(' ').str.get(0)

# 시간단위 부분만 추출
df4['Listed_on_day'] = df4['Listed_on'].str.split(' ').str.get(1)

# time unit -  list of lists
posted = [['second', 'seconds','minute', 'minutes', 'hour', 'hours'], ['day','days'], ['week', 'weeks'], ['month', 'months'], ['year', 'years']]
```

```python
for idx, time in enumerate(posted):
    df4['Listed_on_day'].replace(time, idx, inplace = True)
# 2=week -> 7 days, 3=month -> 30 days, 4=year -> 365
df4['Listed_on_day'].replace(2,7, inplace = True)
df4['Listed_on_day'].replace(3,30, inplace = True)
df4['Listed_on_day'].replace(4,365, inplace = True)
# Listed_on column replacing to days
df4['days_past'] = df4['Listed_on_num'].astype('int64') * df4['Listed_on_day'].astype('int64')
df4['days_past'].tail(5)
```

#### listed_date 변수 처리 - 날짜 계산

```python
# created_at variable(2019-10-26 11:16:00.207427), converting year-month=day
df4.loc[:,'date_created_at'] = pd.to_datetime(df4.loc[:,'created_at'].str[:10])
df4.loc[:,'date_created_at'].head(5)
```

```python
 df4[['date_created_at', "Listed_on", 'days_past']]
```

```python
df4.loc[:,"listed_date"] = df4.loc[:,'date_created_at'] - pd.to_timedelta(df4.loc[:,'days_past'], unit = 'd')
```

```python
 df4[['date_created_at',"Listed_on", 'days_past', 'listed_date']]
```
